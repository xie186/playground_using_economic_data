In this note, I want to see the correlation between Federal funds rate and Unemployment Rate and between Federal funds rate and  National Home Price Index


* Federal Funds Rate and Unemployment Rate: Generally, there is an inverse relationship between the Federal Funds Rate and the Unemployment Rate. When the Federal Funds Rate is low, borrowing and spending are encouraged, which can lead to economic growth and job creation, thereby decreasing the Unemployment Rate. Conversely, when the Federal Funds Rate is high, borrowing and spending are discouraged, which can slow down the economy and lead to job losses, thereby increasing the Unemployment Rate.

  There is a lag between changes in the Federal Funds Rate and changes in the Unemployment Rate. This is because the effects of changes in interest rates on the economy can take time to materialize. For example, when the Federal Reserve raises interest rates to slow down inflation, it may take some time before businesses and consumers respond to the higher borrowing costs by cutting back on spending and investment. This slowdown in economic activity can eventually lead to job losses and a higher Unemployment Rate, but it may take several months or even years for the effects to fully play out.

* Federal Funds Rate and National Home Price Index: There is an inverse relationship between the Federal Funds Rate and the National Home Price Index. When the Federal Funds Rate is low, borrowing and spending are encouraged, which can increase demand for homes and drive up home prices. Conversely, when the Federal Funds Rate is high, borrowing and spending are discouraged, which can decrease demand for homes and drive down home prices.

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [10, 5]
from pandas_datareader import data as pdr
import fix_yahoo_finance as yf
```

```python
def lineplot_FFR_vs_UR(start_date, end_date):
    # Define the tickers for the Federal funds rate and unemployment rate
    tickers = ['FEDFUNDS', 'UNRATE','CSUSHPINSA']

    # Download the data using pandas_datareader and FRED
    yf.pdr_override() # To overcome a bug of pandas_datareader when accessing FRED
    data = pdr.get_data_fred(tickers, start_date, end_date)
    # Plot the data using matplotlib
    fig, ax1 = plt.subplots()
    ax2 = ax1.twinx() # Create a secondary y-axis
    ax1.plot(data.index, data['FEDFUNDS'], label='Federal Funds Rate')
    ax1.plot(data.index, data['UNRATE'], label='Unemployment Rate')
    ax2.plot(data.index, data['CSUSHPINSA'], label='U.S. National Home Price Index', color='green')
    ax1.set_xlabel('Year')
    ax1.set_ylabel('Rate')
    ax2.set_ylabel('Home Price Index')
    #ax1.legend(loc='upper left')
    #ax2.legend(loc='upper right')
    
    # Set the position of the legend
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax2.legend(lines + lines2, labels + labels2, loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=3)

    plt.show()
```

```python
lineplot_FFR_vs_UR('1950-01-01', '2023-01-31')
```

```python
lineplot_FFR_vs_UR('2000-01-01', '2023-01-31')
```

```python
lineplot_FFR_vs_UR('2004-01-01', '2012-01-31')
```
